# 📊 Analisis Data Penjualan (Cleaning & Visualisasi)

Notebook ini berisi proses:
- Penggabungan dataset
- Cleaning data (missing value, duplikasi, outlier)
- Analisis penjualan produk *Pembersih Lantai*
- Visualisasi line chart penjualan per bulan


```python
import pandas as pd
import matplotlib.pyplot as plt

print('Library siap digunakan')
```

```python
# =========================
# LOAD DATA
# =========================

df1 = pd.read_csv('data1.csv')
df2 = pd.read_csv('data2.csv')
df3 = pd.read_csv('data3.csv')

df = pd.concat([df1, df2, df3], ignore_index=True)

print('Total baris setelah gabung:', len(df))
df.head()
```

```python
# =========================
# MISSING VALUE CHECK
# =========================

print(df.isnull().sum())
print('Total missing:', df.isnull().sum().sum())
```

```python
# =========================
# DUPLIKAT CHECK
# =========================

print('Jumlah duplikat:', df.duplicated().sum())

df = df.drop_duplicates()
print('Setelah hapus duplikat:', len(df))
```

```python
# =========================
# OUTLIER REMOVAL (HARGA)
# =========================

Q1 = df['harga'].quantile(0.25)
Q3 = df['harga'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['harga'] >= lower) & (df['harga'] <= upper)]

print('Setelah outlier removal:', len(df))
```

```python
# =========================
# FILTER PRODUK
# =========================

df_produk = df[df['produk'] == 'Pembersih Lantai']
df_produk.head()
```

```python
# =========================
# AGREGASI PER BULAN
# =========================

monthly = df_produk.groupby('bulan')['jumlah_penjualan'].sum().reset_index()
monthly = monthly.sort_values('bulan')
monthly
```

```python
# =========================
# VISUALISASI LINE CHART
# =========================

plt.figure(figsize=(10,5))
plt.plot(monthly['bulan'], monthly['jumlah_penjualan'], marker='o')

plt.title('Penjualan Pembersih Lantai (Bulan 1-12)')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.xticks(range(1,13))
plt.grid(True)

plt.show()
```
